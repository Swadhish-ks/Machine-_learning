```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'Hours' : [1,2,3,4,5,3,2,1,5],
        'Marks' : [30,50,60,70,80,80,50,40,90]}
df =pd.DataFrame(data)
```

```python
df
```

```python
x = df['Hours']
y = df['Marks']

plt.bar(x,y)
plt.show()
```

```python
plt.scatter(df['Hours'],df['Marks'])
```

```python
# independent variable
x = df[['Hours']]
x
```

```python
y = df['Marks']
y
```

## Model building

```python
##import librarie
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
print('intercept (B0)',model.intercept_)
print('Slop (B1)',model.coef_)
```

## Predict

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
pred = model.predict([[10]])
print("predicted marks for 6 hours",pred[0])
```

```python
plt.scatter(x,y,color = 'blue')
plt.plot(x ,model.predict(x),color = 'red')
plt.xlabel ('Hours studied')
plt.ylabel('Mark scored')
plt.title('LINEAR REGRESSION EXAMPLE')
```
