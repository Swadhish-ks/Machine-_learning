### Here city and water source are category type and linear regression models cant use category only support numerics so we need to convert 
## To convert have some method they are
1. ordinal encoding
2. Label encoding
3. one-hot encoding (Dummy variables)

# 1. ordinal encoding

```python
house['Water_Source'].unique()
```

- Here have 'Corporation', 'Borewell', 'Well' so give each category an number manualy like below

```python
house['Water_Source_o'] = house['Water_Source'].map({'Corporation':3 ,'Borewell':2,'Well':1})
```

```python
house['Water_Source'].unique()
```

# 2 Label encoding
- ordinal encoding has some drawback that is large number of datas have ordinal coding is note work so label encoding can be possible
- That is here in label encoding it label automatically

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python

house['water_source_le'] = le.fit_transform(house['Water_Source'])
```

```python
house
```

```python
house['water_source_le'].unique()
```

```python
le.classes_ 
```
